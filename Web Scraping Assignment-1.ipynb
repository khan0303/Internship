```python
import pandas as pd


page = requests.get('https://www.imdb.com/chart/top')
page
```

```python
ratings = []
for x in imdb_rating:
    x = x.get_text().replace('\n', "")
    x = x.strip(" ")
    ratings.append(x)
    
ratings
```

```python
year = soup.find_all('span', class_='secondaryInfo')
year
```

```python
year_of_release = []
for a in year:
    a = a.get_text().replace('\n', "")
    a = a.strip(" ")
    year_of_release.append(a)
    
year_of_release
```

```python
df = pd.DataFrame({'Name' : title, 'Rating' : ratings})
df.head(100)
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup.title.text
```

```python
teams = soup.find_all('span', class_= 'u-hide-phablet')

Team = []
for i in teams:
    Team.append(i.text)
    
print(Team)
```

```python
matches = soup.find('td', class_= 'rankings-block__banner--matches')

Matches = []


Matches.append(matches.text)

print(Matches)
```

```python
match = soup.find_all('td', class_= 'table-body__cell u-center-text')

M = []

for i in match:
    M.append(i.text)
    
print(M)
```

```python
points = soup.find('td', class_= 'rankings-block__banner--points')

Points = []

Points.append(points.text)

print(Points)
```

```python
point = soup.find_all('td', class_= 'table-body__cell u-center-text')

P = []

for i in point:
    P.append(i.text)
    
print(P)
```

```python
Matches = Matches + M[0::2]
Points = Points + P[1::2]
```

```python
teams_ratings = soup.find_all('td', class_= ['table-body__cell u-text-right rating','rankings-block__banner--rating u-text-right'])

ratings = []

for a in teams_ratings:
    a = a.get_text().replace('\n', "")
    ratings.append(a)
    
print(ratings)
```

```python
data = pd.DataFrame({'Team Name': Team[:20], 'Matches': Matches, 'Points': Points, 'Team Rating': ratings})
data
```

```python
name = soup.find_all('a', class_= 'restnt-name ellipsis')
Name = []
for i in name:
    Name.append(i.text)
    
    
print(Name)
```

```python
z = soup.find_all('span', class_= 'double-line-ellipsis')
cuisine = []

for x in z:
    cuisine.append(x.text.split('|')[1])
    
print(cuisine)
```

```python
loc = soup.find_all('div', class_= 'restnt-loc ellipsis')
location = []
for i in loc:
    location.append(i.text)
    
    
print(location)
```

```python
rating = soup.find_all('div', class_= 'restnt-rating rating-4')
ratings = []

for a in rating:
    ratings.append(a.text)
    
print(ratings)
```

```python
img = soup.find_all('img', class_='no-img')
images=[]
for i in img:
    images.append(i['data-src'])
    
    
print(images)
```
